```python
from scipy.optimize import linprog
```

```python
c = [-6, -5]
```

```python
A = [
    [1, 1],     # Milk constraint: x + y <= 5
    [3, 2]      # Choco constraint: 3x + 2y <= 12
]
b = [5, 12]
```

```python
x_bounds = (0, None)
y_bounds = (0, None)
```

```python
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')
```

```python
if result.success:
    print(f"Optimal number of chocolate A (x): {result.x[0]:.2f}")
    print(f"Optimal number of chocolate B (y): {result.x[1]:.2f}")
    print(f"Maximum profit: {-result.fun:.2f}")  # Negative because we minimized -Z
else:
    print("No solution found.")
```

```python
from scipy.optimize import linprog
```

```python
c = [-50, -120]
```

```python
A = [
    [7000, 2000],
    [10, 30]
]
b = [700000, 1200]
a_1=[[1,1]]
b_1=[100]
```

```python
x_bounds = (0, None)
y_bounds = (0, None)
```

```python
result = linprog(c, A_ub=A, b_ub=b, A_eq=a_1, b_eq=b_1, bounds=[x_bounds, y_bounds], method='highs')
```

```python
if result.success:
    print(f"Optimal hectares for Wheat (x): {result.x[0]:.2f}")
    print(f"Optimal hectares for Barley (y): {result.x[1]:.2f}")
    print(f"Maximum profit: {-result.fun:.2f}")  # Negative because we minimized -Z
else:
    print("No solution found.")
```
